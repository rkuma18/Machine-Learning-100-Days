3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(x='Sex', y='Age', hue='Survived', data=titanic)
```

4. Distplot (Numerical - Categorical)

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(data=titanic[titanic['Survived'] == 0], x='Age', label='Did Not Survive')
sns.kdeplot(data=titanic[titanic['Survived'] == 1], x='Age', label='Survived')
plt.legend()
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

6. ClusterMap (Categorical - Categorical)

```python
ct = pd.crosstab(titanic['Parch'], titanic['Survived'])
sns.clustermap(ct, annot=True, cmap='Blues')
plt.show()
```
