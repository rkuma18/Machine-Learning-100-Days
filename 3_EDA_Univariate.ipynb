```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
```

```python
df.head()
```

1. Categorical Data

```python
# a. Countplot
sns.countplot(x='Embarked', data=df)
plt.show()
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
# b. PieChart
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

2. Numerical Data

```python
# a. Histogram

plt.hist(df['Age'],bins=5)
```

```python
# b. Distplot

sns.histplot(df['Age'], kde=True, stat='percent')  # Normalizes to percentage
plt.title("Age Distribution (%)")
plt.xlabel("Age")
plt.ylabel("Percentage")
plt.show()
```

```python
# c. boxplot
sns.boxplot(df['Age'])
```
